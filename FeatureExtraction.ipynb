```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv('anime.csv')
```

```python
df.loc[2]['Title']
```

```python
def extract_text(text):
  check=False
  texts=""
  for i in text:
    if i == ')':
      return texts

    if check==True:

      texts=texts+i
    if i=='(':

      check=True
  return texts
```

```python
df['Episodes']=df['Title'].apply(extract_text)
```

```python
df
```

```python
df['Episodes']=df['Episodes'].str.replace(' eps','')
```

```python
df
```

```python
df['Episodes']=df['Episodes'].astype('int')
df
```

```python
df.loc[1]['Title']
```

```python
def extract_time(txt):
  # check=False
  data=""
  for i in range(len(txt)):
    if txt[i]==")":
      for j in range(i+1,i+20):
        data=data+txt[j]
  return data
```

```python
df['Total Time']=df['Title'].apply(extract_time)
```

```python
df
```

```python
def total_months(date_range):
    start, end = date_range.split(" - ")

    start_date = datetime.strptime(start, "%b %Y")
    end_date = datetime.strptime(end, "%b %Y")

    return (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month) + 1
```

```python
df["Total Month"] = df["Total Time"].apply(total_months)
df
```
